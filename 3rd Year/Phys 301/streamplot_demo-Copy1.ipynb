```python
import numpy as np
import matplotlib.pyplot as plt

def func(m,n):
    k = 8.99e9
    Z, X = np.mgrid[-m:m:100j,-m:m:100j]
    def Ef(L):
        x = L[0]
        z = L[1]
        q = L[2]
        
        ex = k*(X-x)*q/((X-x)**2+(Z-z)**2)**(3/2)
        ez = k*(Z-z)*q/((X-x)**2+(Z-z)**2)**(3/2)
        return (ex,ez)
    
    U = Ef([1,0,-1])[0]+Ef([-1,0,-1])[0]+Ef([0,-2,-1])[0]+Ef([0,2,3])[0]
    V = Ef([1,0,-1])[1]+Ef([-1,0,-1])[1]+Ef([0,-2,-1])[1]+Ef([0,2,3])[1]
    
    plt.figure(n)
    plt.title('Electric Vector Field of the 4 point charges')
    plt.streamplot(X,Z,U,V,density=[2,2])
    plt.plot(1,0,'bo')
    plt.plot(-1,0,'bo')
    plt.plot(0,2,'ro')
    plt.plot(0,-2,'bo')
    plt.savefig('{}x{}.pdf'.format(-m,m))
    plt.show()
    
func(4,1)
func(10,2)

    
```
