```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 5a)I'll use meshgrid and mgrid both to do this question, plotting a color image
# and a contour plot. They'll give same results. 
```

```python
def func(S0,a,v):
    Sv = S0*(v/400)**a
    return Sv
```

```python
alphas = np.linspace(-3,3,200)
freqs = np.linspace(300,501,200)
a, v = np.meshgrid(alphas,freqs)

result = func(3,a,v)
plt.imshow(result, cmap ='afmhot',origin='lower')
plt.colorbar()
```

```python
plt.figure(figsize=(15,10))
S0 = 3 #units of mJy
a, v = np.mgrid[-3:4, 300:501]
Sv = S0*(v/400)**a
plt.contourf(a,v,Sv,cmap ='afmhot')
plt.colorbar()
plt.xlabel(r'Spectral Index,$\alpha$',fontsize=15)
plt.ylabel(r'Frequency, $\nu$',fontsize=15)
plt.title(r'Flux Density, $S(\nu)$, of radio source in mJy',fontsize=20)
plt.savefig('404-Hw1Q5a.pdf')
```

```python
plt.figure(figsize=(15,10))
S0 = 3E-29
a = np.linspace(-3,3,300)
num = 500**(a+1)-300**(a+1)
den = (400**a)*(a+1)
E = S0*np.pi*2500*num/den
plt.plot(a,E)
plt.xlabel(r'Spectral Index, $\alpha$',fontsize=15)
plt.ylabel(r'Energy, $J$',fontsize=15)
plt.title('Energy received by telescope in 1s',fontsize=20)
```
