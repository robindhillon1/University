```python
from scipy.stats import linregress
import numpy as np
import matplotlib.pyplot as plt
```

# Q3
Once the slope is found, a linear model is fitted to the data as shown in red. 

```python
wv = [1,1.1,1.2,1.3,1.4,1.5]
flux = [-339.12,-336.294,-333.468,-330.642,-328.13,-325.618]

result = linregress(wv, flux)
print(result)
model = result[0]*np.asarray(wv)+result[1]

plt.figure(figsize=(12,10))
plt.plot(wv,flux,'ko-',label="Data")
plt.plot(wv, model,'ro--',label="Model")
plt.xlabel(r"Water Vapor Factor, $WV_{fac}$",fontsize=15)
plt.ylabel(r"Flux $(W/m^2)$",fontsize=15)
plt.title("Upward IR Flux as a function of Water Vapor Factor",fontsize=18)
plt.legend()
plt.grid()
```

```python
e0 = np.asarray([1900,2100,2300,2500,2700,2900])  # Pa

H, Rv, T = 5000, 461, 285
M = (H*e0)/(Rv*T)  # kg/m^2
Mv = M/18  # kmoles/m^2
result2 = linregress(Mv, flux)
print(result2)

model2 = result2[0]*Mv+result2[1]

plt.figure(figsize=(12,10))
plt.plot(Mv,flux,'ko-',label="Data")
plt.plot(Mv, model2,'ro--',label="Model")
plt.xlabel("$H_2O$ Kmoles/$m^2$",fontsize=15)
plt.ylabel("Flux $(W/m^2)$",fontsize=15)
plt.title("Upward IR Flux as a function of $H_2O$ Kmoles/$m^2$",fontsize=18)
plt.legend()
plt.grid()
```

# Q4

```python
print(r"Slope: {:.4f} (W/m^2)/(kmoles/m^2)".format(result2[0]))
```

For clarity, we can invert the y axis as it clearly shows that as the concentrations of $H_2O$ are increased, the upward flux (hence the negative sign) decreases. This means that more flux is trapped with increasing concentrations of $H_2O$.

```python
plt.figure(figsize=(12,10))
plt.plot(Mv,flux,'ko-',label="Data")
plt.plot(Mv, model2,'ro--',label="Model")
plt.xlabel("$H_2O$ Kmoles/$m^2$",fontsize=15)
plt.ylabel("Flux $(W/m^2)$",fontsize=15)
plt.gca().invert_yaxis()
plt.title("Upward IR Flux as a function of $H_2O$ Kmoles/$m^2$",fontsize=18)
plt.legend()
plt.grid()
```

# Q5

```python
CO2 = np.asarray([400, 440, 480, 520, 560, 600])
flux3 = [-400.664, -399.722, -399.408, -398.78, -398.152, -397.838]

result3 = linregress(CO2, flux3)
print(result3)
model3 = result3[0]*CO2+result3[1]

plt.figure(figsize=(12,10))
plt.plot(CO2,flux3,'ko-',label="Data")
plt.plot(CO2, model3,'ro--',label="Model")
plt.xlabel("$CO_2$ mixing ratios",fontsize=15)
plt.ylabel(r"Flux $(W/m^2)$",fontsize=15)
plt.title("Upward IR Flux as a function of $CO_2$ Mixing Ratios",fontsize=18)
plt.legend()
plt.grid()
```

```python
Matm = 10000/29  # kmol/m^2
MCO2 = Matm*CO2*1E-6  # CO2 kmol/m^2 
result5 = linregress(MCO2, flux3)
print(result5)
model5 = result5[0]*MCO2+result5[1]

plt.figure(figsize=(12,10))
plt.plot(MCO2,flux3,'ko-',label="Data")
plt.plot(MCO2, model5,'ro--',label="Model")
plt.xlabel("$CO_2$ Kmoles/$m^2$",fontsize=15)
plt.ylabel(r"Flux $(W/m^2)$",fontsize=15)
plt.title("Upward IR Flux as a function of $CO_2$ Kmoles/$m^2$",fontsize=18)
plt.legend()
plt.grid()
```

Based on the answers above, $CO_2$ is the more potent greenhouse gas per molecule as it traps more flux than $H_2O$ as seen from the slopes. From Q4, we see that as the water vapor factor increases, flux is trapped at a rate of 6.4075 $(W/m^2)/(kmoles/m^2)$; in other words, every $kmoles/m^2$ of $H_2O$ traps a flux of 6.4075 $(W/m^2)$.
